```python
START = 0
END = 500
array = []
with open('results.txt') as f:
    for line in f: # read rest of lines
        array.append([float(x) for x in line.split()])


opt_array = []
with open('optimized_results.txt') as f:
    for line in f: # read rest of lines
        opt_array.append([float(x) for x in line.split()])

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
array = np.array(array)
opt_array = np.array(opt_array)
poses = pd.read_csv('00.txt', delimiter=' ', header=None)
gt = np.zeros((len(poses), 3, 4))
for i in range(len(poses)):
      gt[i] = np.array(poses.iloc[i]).reshape((3, 4))

array = array[:END]
gt = np.array(gt)
gt_ = gt[START:START + len(array)]
print(gt_[0].shape)
for i in array:
  i += gt_[0][:,3]

xs = gt_[:, 0, 3]
ys = gt_[:, 1, 3]
zs = gt_[:, 2, 3]

fig = plt.figure(figsize = (16,8))
ax = plt.axes(projection='3d')
ax.plot(array[:len(array),0], array[:len(array),1], array[:len(array),2], c='chartreuse')
ax.plot(opt_array[:len(opt_array),0], opt_array[:len(opt_array),1], opt_array[:len(opt_array),2], c='blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(0, -90)
ax.plot(xs, ys, zs, c='blue')
```
